### 4. Get rid of useless columns for the Business View

```python
bv = bv.loc(axis=1)["benefit_type", "benefit_name", "benefit_period", "Concatener", "Traite", "Section", "Code Garantie", "extract_date", "qs_amount", "claim_event_date"]
```

```python
bv.tail()
```

### Data treatments : formating dates, creating new columns like Survenance, Devise, Code cedante depending on the ceding company
* Check the PREPAWEB files to see how it's done 

```python
for col in bv.columns :
    if "date" in col :
        bv[col] = pd.to_datetime(bv[col])
        
        
bv['Survenance']= bv['claim_event_date'].dt.year

SurvenanceisNan = bv.loc[np.isnan(bv["Survenance"])].shape[0]
print("Numbers of rows without Claim Event Date : ", SurvenanceisNan, "rows")        

# using list comprehension
# checking if filename in Australian Treaties

test_string = os.path.basename(filepath)
australia = ["AIA", "mlc", "CareSuper", "TAL"]
res = [ele for ele in australia if(ele in test_string)]
  
if res :
    bv['Devise'] = "AUD"
    if 'mlc' in res : 
        bv['Code cedante']= 365040
    if 'AIA' in res :
        bv['Code cedante']= 359257

bv
```

### Grouping rows that have the same characteristics and summing their amount

```python
bv_final = bv.groupby(["Traite", "Section", "Code Garantie", "extract_date", "Survenance", "Code cedante"]) \
            .agg({'qs_amount': 'sum'}) \
            .reset_index()
```

```python
bv_final.fillna(0)
bv_final["Code Garantie"] = bv_final["Code Garantie"].astype(int)
bv_final["Survenance"] = bv_final["Survenance"].astype(int)
```

```python
bv_final['Debut'] = bv_final['extract_date'].to_numpy().astype('datetime64[M]')

if "qs_amount" in bv_final.columns :
    bv_final["Nature"] = "210"
    
bv_final.rename(columns={'qs_amount': 'Montant', 
                           'extract_date': 'Fin'}, inplace=True)
```

#### Reordering the columns

```python
order = ['Code cedante', 'Debut', 'Fin', 'Traite', 'Section', 'Nature' , 'Code Garantie' , 'Survenance' , 'Montant' , 'Devise']
```

```python
bv_final = bv_final.reindex(order, axis ='columns').sort_values(by=['Fin'], ascending = False)
```

```python
for colu in bv_final.columns :
    if colu == "Fin" or colu == "Debut" :
        bv_final[colu] = bv_final[colu].dt.strftime('%d/%m/%Y')
```

```python
bv_final['Montant'] = bv_final['Montant'].round(decimals= 2)
```
